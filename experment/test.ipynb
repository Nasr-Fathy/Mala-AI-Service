```python
import os
import json
import google.generativeai as genai

# 1) API key
genai.configure(api_key=os.environ["GOOGLE_AI_API_KEY"])

# 2) model
model = genai.GenerativeModel(os.getenv("GOOGLE_AI_MODEL", "gemini-3-flash-preview"))

# 3) request
prompt = """
Return ONLY valid JSON in this format:
{
  "message": "hello",
  "numbers": [1, 2, 3]
}
"""

response = model.generate_content(
    prompt,
    generation_config=genai.GenerationConfig(
        max_output_tokens=1024,
        temperature=0.1,
        response_mime_type="application/json",
    ),
)

# 4) print raw response text
print("=== RAW TEXT ===")
print(response.candidates[0].content.parts[0].text)

# 5) usage metadata
um = response.usage_metadata

print("\n=== USAGE METADATA ===")
fields = [
    "prompt_token_count",
    "cached_content_token_count",
    "candidates_token_count",
    "tool_use_prompt_token_count",
    "thoughts_token_count",
    "total_token_count",
    "prompt_tokens_details",
    "cache_tokens_details",
    "candidates_tokens_details",
    "tool_use_prompt_tokens_details",
]

for f in fields:
    print(f"{f}: {getattr(um, f, None)}")

# 6) fallback introspection: print any extra attrs not listed above
print("\n=== ALL NON-CALLABLE ATTRS ===")
for name in dir(um):
    if not name.startswith("_"):
        try:
            value = getattr(um, name)
            if not callable(value):
                print(f"{name}: {value}")
        except Exception:
            pass
```

```python
from pathlib import Path
from pypdf import PdfReader, PdfWriter


OUTPUT_DIR = Path(r"C:\Users\TF-026\OneDrive\Desktop\backend-main\ai-service\Test doc\Doc after splite")


def extract_pages_to_pdf(pdf_path: str, page_numbers: list[int], output_name: str | None = None) -> str:
    """
    Extract selected pages from a PDF and save them as a new PDF.

    Args:
        pdf_path: Full path to source PDF.
        page_numbers: List of 1-indexed page numbers to extract, e.g. [1, 3, 5]
        output_name: Optional output file name. If None, auto-generated.

    Returns:
        Full path of the saved PDF.
    """
    pdf_path = Path(pdf_path)
    OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

    if not pdf_path.exists():
        raise FileNotFoundError(f"PDF not found: {pdf_path}")

    if not page_numbers:
        raise ValueError("page_numbers cannot be empty")

    reader = PdfReader(str(pdf_path))
    writer = PdfWriter()

    total_pages = len(reader.pages)

    # remove duplicates while preserving order
    seen = set()
    cleaned_pages = []
    for p in page_numbers:
        if not isinstance(p, int):
            raise TypeError(f"All page numbers must be integers. Got: {p}")
        if p < 1 or p > total_pages:
            raise ValueError(f"Page number {p} is out of range. PDF has {total_pages} pages.")
        if p not in seen:
            seen.add(p)
            cleaned_pages.append(p)

    for page_num in cleaned_pages:
        writer.add_page(reader.pages[page_num - 1])  # convert 1-indexed to 0-indexed

    if output_name is None:
        pages_part = "_".join(map(str, cleaned_pages))
        output_name = f"{pdf_path.stem}_pages_{pages_part}.pdf"

    output_path = OUTPUT_DIR / output_name

    with open(output_path, "wb") as f:
        writer.write(f)

    return str(output_path)
```

```python
pdf_path = r"C:\Users\TF-026\OneDrive\Desktop\backend-main\ai-service\Test doc\ميزانية شركة امواج الخير 2024م.pdf"
pages = [3,4]

new_pdf_path = extract_pages_to_pdf(pdf_path, pages)
print("Saved to:", new_pdf_path)
```
